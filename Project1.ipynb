```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Removing the first empty row from dataset and setting the index without any header

laliga = pd.read_csv('Laliga.csv',  skiprows =1)
laliga.head()
```

```python
# 1)  replace dashes with 0 
laliga.replace('-',0, inplace =True)
laliga.head()
```

```python
# Converting required columns datatypes as integer before starting on calculations

#laliga.astype({'GamesPlayed': 'int64'}).dtypes
#laliga.astype("int64", errors="ignore")
laliga=laliga.apply(pd.to_numeric, errors='ignore')
laliga.dtypes
```

```python
#2) Print all the teams which have started playing between 1930-1980
#Debut is the column which will identify when team started playing. 

#print((laliga['Debut'] >'1930') & (laliga['Debut'] <'1980' ))
#laliga.loc[:,'Debut']
teamlist = laliga[(laliga['Debut'] >'1930') & (laliga['Debut'] <'1980' )]
print(teamlist['Team'])
teamlist['Team'].count()
```

```python
#3) list of teams which came Top 5 in terms of points

laliga.sort_values('Points' ,ascending = False).head(5)
```

```python
#4) Creating function goal difference and printing team name
# Method 2 to implement the goal difference function

def Goal_diff_count():
    for i in range(len(laliga)):

#Created new column for finding the difference
        laliga['GoalDiff'] = laliga['GoalsFor']- laliga['GoalsAgainst']  
    
    temp = laliga[['Team','GoalDiff']]
    df1 = laliga.iloc[laliga.GoalDiff.idxmax(),1]
    df2 = laliga.iloc[laliga.GoalDiff.idxmin(),1]

    return temp, df1,df2;

#Calling the function and getting the teamnames with goal difference, team with maximum goal difference and minimum difference
tempdf, maxdf, mindf= Goal_diff_count()
print(tempdf)
print(maxdf)
print(mindf)
    

    
```

```python
# 5) Create a new column with name “Winning Percent"
laliga['WinningPercent'] = (laliga['GamesWon']/laliga['GamesPlayed'])*100

# Replacing 0 for Nan values 
laliga.fillna(0, inplace =True)

# Setting up decimal point to 2places
laliga['WinningPercent'] = laliga['WinningPercent'].round(decimals = 2)
print(laliga[['Team','WinningPercent']])

# Printing top 5 teams on the basis of winning percentage
laliga.sort_values('WinningPercent' ,ascending = False).head(5)
```
